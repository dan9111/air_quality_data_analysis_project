### Cleaning Data

Seluruh tabel memiliki dua problem yang teridentifikasi yaitu:
- Memiliki nilai `null`
- Memiliki kemungkinan nilai noise pada indeks polutan

Untuk indeks PM2.5, PM10, SO2, NO2, CO, dan O3 dapat dilihat baris-baris disekitar nilai yang diduga outlier atau noise, apabila masuk dalam pola maka indeks tersebut bukanlah invalid value

##### Daerah Aotizhongxin

```python
max_index = air_aotizhongxin_df.index[air_aotizhongxin_df['PM2.5'] == air_aotizhongxin_df['PM2.5'].max()].tolist()
print("Maximum PM2.5 count:", len(max_index))
print("PM2.5:")
print(air_aotizhongxin_df.loc[max_index[0] - 4:max_index[0] + 4]['PM2.5'])
print('')
max_index = air_aotizhongxin_df.index[air_aotizhongxin_df['PM10'] == air_aotizhongxin_df['PM10'].max()].tolist()
print("Maximum PM10 count:", len(max_index))
print("PM10:")
print(air_aotizhongxin_df.loc[max_index[0] - 4:max_index[0] + 4]['PM10'])
print('')
max_index = air_aotizhongxin_df.index[air_aotizhongxin_df['SO2'] == air_aotizhongxin_df['SO2'].max()].tolist()
print("Maximum SO2 count:", len(max_index))
print("SO2:")
print(air_aotizhongxin_df.loc[max_index[0] - 4:max_index[0] + 4]['SO2'])
print('')
max_index = air_aotizhongxin_df.index[air_aotizhongxin_df['NO2'] == air_aotizhongxin_df['NO2'].max()].tolist()
print("Maximum NO2 count:", len(max_index))
print("NO2:")
print(air_aotizhongxin_df.loc[max_index[0] - 4:max_index[0] + 4]['NO2'])
print('')
max_index = air_aotizhongxin_df.index[air_aotizhongxin_df['CO'] == air_aotizhongxin_df['CO'].max()].tolist()
print("Maximum CO count:", len(max_index))
print("CO:")
print(air_aotizhongxin_df.loc[max_index[4] - 4:max_index[4] + 4]['CO'])
print('')
max_index = air_aotizhongxin_df.index[air_aotizhongxin_df['O3'] == air_aotizhongxin_df['O3'].max()].tolist()
print("Maximum O3 count:", len(max_index))
print("O3:")
print(air_aotizhongxin_df.loc[max_index[0] - 4:max_index[0] + 4]['O3'])
```

```python
all_stations = air_all_df['station'].unique()
fig, ax = plt.subplots(nrows=6, ncols=2, figsize=(20, 35))
for i in range(len(all_stations)):
    rows = i // 2
    cols = i % 2
    sns.lineplot(
        x='year',
        y='PM2.5',
        data=yearly_air_quality_df.loc[all_stations[i]],
        color='red',
        marker='o',
        linestyle='solid',
        label='PM2.5',
        ax=ax[rows, cols]
    )
    sns.lineplot(
        x='year',
        y='PM10',
        data=yearly_air_quality_df.loc[all_stations[i]],
        color='orange',
        marker='o',
        linestyle='solid',
        label='PM10',
        ax=ax[rows, cols]
    )
    ax[rows, cols].set_title(f"{all_stations[i]} Air Pollution from 2013 to 2017", fontsize=15, loc="center")
    ax[rows, cols].set_xlabel(None)
    ax[rows, cols].set_ylabel('PM2.5/PM10', fontsize=12)
    ax[rows, cols].set_xticks(yearly_air_quality_df.loc[all_stations[i]].index)
```

Setiap daerah memiliki puncak polusi udara pada tahun 2014, setiap daerah juga mengalami tingkat polusi terendah pada tahun 2016, lalu naik kembali pada tahun 2017

### Pertanyaan 2: Bagaimana hubungan antara kualitas udara dengan suhu udara?

Untuk menentukan hubungan kualitas udara dan suhu udara dapat menggunakan correlation heatmap

```python
fig,ax = plt.subplots(figsize=(12, 10))
corr = air_all_df.select_dtypes('number').corr()
sns.heatmap(corr, annot=True, fmt=".2f", ax=ax, cmap='coolwarm')
ax.set_title(f"Correlation Heatmap", fontsize=15, loc="center")
ax.set_xlabel(None)
ax.set_ylabel(None)
```

Jika dilihat dari parameter kualitas udara, Suhu udara tidak dipengaruhi oleh indeks kualitas udara dan komposisi polutannya (SO2, NO2, dan CO), namun komposisi O3 berpengaruh pada suhu udara jika dilihat korelasinya (berbanding lurus), hal ini karena O3 (ozone) merupakan gas rumah kaca yang efektif dalam memperangkap sinar matahari, sehingga banyaknya kadar ozone akan berpengaruh pada suhu udara.

### Pertanyaan 3: Daerah mana yang paling bersih polusi dan yang paling terpolusi?

Perlu mengelompokkan tingkat polusi dari tiap daerah (station)

```python
bystation_df = air_all_df.groupby(by='station').agg({
    'PM2.5': 'mean',
    'PM10': 'mean',
})
```

```python
bystation_df_pm25 = bystation_df.sort_values(by='PM2.5', ascending=False)
bystation_df_pm10 = bystation_df.sort_values(by='PM10', ascending=False)
```

Lalu data polusi per daerah akan dibagi menjadi dua yaitu untuk tingkat PM2.5 dan PM10

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
colors = ['#C01919','#F4BB94','#F4BB94','#F4BB94','#F4BB94','#F4BB94','#F4BB94','#F4BB94','#F4BB94','#F4BB94','#F4BB94','#78B21F']
sns.barplot(x=bystation_df_pm25.index, y='PM2.5', data=bystation_df_pm25, ax=ax[0], palette=colors)
sns.barplot(x=bystation_df_pm10.index, y='PM10', data=bystation_df_pm10, ax=ax[1], palette=colors)
ax[0].set_title("Average PM2.5 by Station", fontsize=15, loc="center")
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=45)
ax[0].tick_params(axis="y", labelsize=12)
ax[1].set_title("Average PM10 by Station", fontsize=15, loc="center")
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=45)
ax[1].tick_params(axis="y", labelsize=12)
plt.show()
```

Dingling adalah daerah yang paling bersih polusi dari semua daerah. Untuk daerah yang paling banyak polusi, Gucheng menduduki peringkat pertama untuk tingkat PM10, dan Dongsi menduduki peringkat pertama untuk tingkat PM2.5

Jika dibandingkan mana yang lebih membahayakan, maka PM2.5 adalah yang paling bahaya, karena PM2.5 (partikel 2.5 micron) mudah untuk masuk dalam paru-paru manusia sehingga lebih invasif.
